## Heatmap 

```python
heat_map = pd.DataFrame({'idx': energy_data['date'], 'date': energy_data['date'], 
                         'Appliances': energy_data['Appliances']})
heat_map.date = heat_map.date.astype('str')
heat_map.idx = heat_map.idx.astype('datetime64[ns]')
heat_map.set_index('idx', inplace = True)
```

```python
def heatGen(data):
    dic = dict()
    days_lst = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
    time_lst = [x for x in range(0,24)] 
    for i in range(0, len(time_lst)): 
        dic[i] = {}
    
    for i in range(0, len(time_lst)): 
        for x in days_lst:
            dic[i][x] = 0
            
    for i in range(0, data.shape[0]): 
         dic[datetime.strptime(data.iloc[i]['date'], "%Y-%m-%d %H:%M:%S").hour][days_lst[datetime.strptime(data.iloc[i]['date'], "%Y-%m-%d %H:%M:%S").weekday()]] += (data.iloc[i]['Appliances'])               
    return dic                                                                        
```

```python
w1 = heatGen(heat_map['2016-01-11' : '2016-01-17'])
w2 = heatGen(heat_map['2016-01-18' : '2016-01-24'])
w3 = heatGen(heat_map['2016-01-25' : '2016-01-31'])
w4 = heatGen(heat_map['2016-02-01' : '2016-02-07'])
```

```python
heat_w1 = pd.DataFrame({'Saturday': [w1[x]['Saturday'] for x in w1], 
                        'Sunday': [w1[x]['Sunday'] for x in w1],
                        'Monday': [w1[x]['Monday'] for x in w1],
                        'Tuesday': [w1[x]['Tuesday'] for x in w1],
                        'Wednesday': [w1[x]['Wednesday'] for x in w1],
                        'Thursday': [w1[x]['Thursday'] for x in w1],
                        'Friday': [w1[x]['Friday'] for x in w1],}) 

heat_w2 = pd.DataFrame({'Saturday': [w2[x]['Saturday'] for x in w2], 
                        'Sunday': [w2[x]['Sunday'] for x in w2],
                        'Monday': [w2[x]['Monday'] for x in w2],
                        'Tuesday': [w2[x]['Tuesday'] for x in w2],
                        'Wednesday': [w2[x]['Wednesday'] for x in w2],
                        'Thursday': [w2[x]['Thursday'] for x in w2],
                        'Friday': [w2[x]['Friday'] for x in w2],})

heat_w3 = pd.DataFrame({'Saturday': [w3[x]['Saturday'] for x in w3], 
                        'Sunday': [w3[x]['Sunday'] for x in w3],
                        'Monday': [w3[x]['Monday'] for x in w3],
                        'Tuesday': [w3[x]['Tuesday'] for x in w3],
                        'Wednesday': [w3[x]['Wednesday'] for x in w3],
                        'Thursday': [w3[x]['Thursday'] for x in w3],
                        'Friday': [w3[x]['Friday'] for x in w3],}) 

heat_w4 = pd.DataFrame({'Saturday': [w4[x]['Saturday'] for x in w4], 
                        'Sunday': [w4[x]['Sunday'] for x in w4],
                        'Monday': [w4[x]['Monday'] for x in w4],
                        'Tuesday': [w4[x]['Tuesday'] for x in w4],
                        'Wednesday': [w4[x]['Wednesday'] for x in w4],
                        'Thursday': [w4[x]['Thursday'] for x in w4],
                        'Friday': [w4[x]['Friday'] for x in w4],}) 
```

## Using dummies days of week

```python
test1 = testing_data.drop(['date', 'Appliances', 'rv1', 'rv2', 'WeekStatus', 'Day_of_week', 'Visibility'], axis = 1) 
train1 = training_data.drop(['date', 'Appliances', 'rv1', 'rv2', 'WeekStatus', 'Day_of_week', 'Visibility'], axis = 1) 

day_week = training_data['Day_of_week'] 
train1_days = train1.join(pd.get_dummies(day_week))

day_week = testing_data['Day_of_week'] 
test1_days = test1.join(pd.get_dummies(day_week))
 
```

```python
week_status = testing_data['WeekStatus'] 
test1_full = test1_days.join(pd.get_dummies(week_status))

week_status = training_data['WeekStatus'] 
train1_full = train1_days.join(pd.get_dummies(week_status))
```

## Regression Model 

```python
import numpy as np 

def MAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true),np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred)/y_true))*100
```

```python
reg = LinearRegression()
y_train = training_data[['Appliances']]
x_train = train1_full 
reg.fit(x_train, y_train)

y_train_pred = reg.predict(x_train)
print("RMSE: ", sqrt(mean_squared_error(y_train, y_train_pred)))
print("R squared: ", r2_score(y_train, y_train_pred))
print("MAE: ", mean_absolute_error(y_train, y_train_pred))
print("MAPE: ", MAPE(y_train, y_train_pred))
```

```python
y_test = testing_data[['Appliances']]
x_test = test1_full 
y_test_pred = reg.predict(x_test)

r2 = r2_score(y_test, y_test_pred) 

print("RMSE: ", sqrt(mean_squared_error(y_test, y_test_pred)))
print("R squared: ", r2_score(y_test, y_test_pred))
print("MAE: ", mean_absolute_error(y_test, y_test_pred))
print("MAPE: ", MAPE(y_test, y_test_pred))
```

```python
estimator = LinearRegression()
RMSEs = []
for i in range(1, 35):
# for i in range(1, 10):
    selector = RFE(estimator, i)
    selector = selector.fit(x_train, y_train)  
    supp = selector.get_support()

    predicted = selector.predict(x_test)
    RMSE = sqrt(mean_squared_error(y_test, predicted))
    RMSEs.append(RMSE)

best = 1
for item in RMSEs:
    if item < RMSEs[best - 1]:
        best = RMSEs.index(item) + 1

plt.grid(alpha=0.5)  
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('Variables')
plt.ylabel('RMSE (cross-validation)')
plt.plot(range(1, 35), RMSEs, marker = 'o', color = 'blue', markeredgewidth = 1 ,markeredgecolor = 'darkblue', markerfacecolor = 'None')
plt.plot(best, RMSEs[best-1], marker = 'o', markerfacecolor = 'lightblue')
```
